```python
import numpy as np
from sklearn.feature_selection import mutual_info_regression
df['new_feature'] = (df['HIGH_CREDIT_OR_SANCTIONED_AMOUNT']/(df['MONTH_OPENED']+1))
non_linear_corr = mutual_info_regression(df[['new_feature']], df['ACTUAL_ROI']).item()
linear_corr = df[['new_feature', 'ACTUAL_ROI']].corr().iloc[0, 1]
print(linear_corr)
print(non_linear_corr)
```

```python
import numpy as np
from sklearn.feature_selection import mutual_info_regression
correlation_before = df[['HIGH_CREDIT_OR_SANCTIONED_AMOUNT', 'ACTUAL_ROI']].corr().iloc[0, 1]

mi_before_log_transform = mutual_info_regression(df[['HIGH_CREDIT_OR_SANCTIONED_AMOUNT']], df['ACTUAL_ROI']).item()

df['log_high_credit'] = np.log1p(df['HIGH_CREDIT_OR_SANCTIONED_AMOUNT'])

correlation_after = df[['log_high_credit', 'ACTUAL_ROI']].corr().iloc[0, 1]

mi_after_log_transform = mutual_info_regression(df[['log_high_credit']], df['ACTUAL_ROI']).item()
print(f'Correlation before logarithmic transformation: {correlation_before:.4f}')
print(f'Mutual Information before logarithmic transformation: {mi_before_log_transform:.4f}\n')

print(f'Correlation after logarithmic transformation: {correlation_after:.4f}')
print(f'Mutual Information after logarithmic transformation: {mi_after_log_transform:.4f}')
```
