```python
df=pd.read_csv('train_main_loan.csv')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming your DataFrame is named 'df'

# Exclude rows with NaN values in REPAYMENT_TENURE
df_no_nan = df.dropna(subset=['REPAYMENT_TENURE'])

# Calculate average REPAYMENT_TENURE for each category of OWNERSHIP_TYPE
ownership_type_avg_tenure = df_no_nan.groupby('OWNERSHIP_TYPE')['REPAYMENT_TENURE'].mean().reset_index()

# Plot average REPAYMENT_TENURE against OWNERSHIP_TYPE
plt.figure(figsize=(12, 6))
sns.barplot(x='OWNERSHIP_TYPE', y='REPAYMENT_TENURE', data=ownership_type_avg_tenure)
plt.title('Average REPAYMENT_TENURE by OWNERSHIP_TYPE')
plt.show()

# Calculate average REPAYMENT_TENURE for each category of OCCUPATION_TYPE
occupation_type_avg_tenure = df_no_nan.groupby('OCCUPATION_TYPE')['REPAYMENT_TENURE'].mean().reset_index()

# Plot average REPAYMENT_TENURE against OCCUPATION_TYPE
plt.figure(figsize=(12, 6))
sns.barplot(x='OCCUPATION_TYPE', y='REPAYMENT_TENURE', data=occupation_type_avg_tenure)
plt.title('Average REPAYMENT_TENURE by OCCUPATION_TYPE')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Assuming you have the DATE_OPENED and ROI data in your DataFrame
# For example, if you have a DataFrame called df:
# df = pd.read_csv('your_data.csv')  # Make sure to load your actual data

# Convert DATE_OPENED to datetime format
df['DATE_OPENED'] = pd.to_datetime(df['DATE_OPENED'], format='%d-%m-%Y')

# Extract the year from DATE_OPENED
df['Year'] = df['DATE_OPENED'].dt.year

# Group by year and calculate the average ROI
avg_roi_by_year = df.groupby('Year')['ACTUAL_ROI'].mean()

# Plot the results
plt.figure(figsize=(10, 6))
avg_roi_by_year.plot(kind='bar', color='skyblue')
plt.title('Average ROI Year-wise')
plt.xlabel('Year')
plt.ylabel('Average ROI')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Assuming you have the TU_SCORE and LOAN_CLASSIFICATION data in your DataFrame
# For example, if you have a DataFrame called df:
# df = pd.read_csv('your_data.csv')  # Make sure to load your actual data

# Create a new column for aggregated LOAN_CLASSIFICATION
bins = [0, 50, 100, 150, 200, 250, 300, 350, 400, 500]
labels = ['1-50', '51-100', '101-150', '151-200', '201-250', '251-300', '301-350', '351-400', '401-500']
df['LOAN_CLASS_SECTION'] = pd.cut(df['LOAN_CLASSIFICATION'], bins=bins, labels=labels, include_lowest=True)

# Plot box plot
plt.figure(figsize=(12, 8))
sns.boxplot(x='LOAN_CLASS_SECTION', y='TU_SCORE', data=df)
plt.title('TU_SCORE vs. LOAN_CLASSIFICATION (Aggregated Sections)')
plt.xlabel('LOAN_CLASS_SECTION')
plt.ylabel('TU_SCORE')
plt.show()
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have the TU_SCORE, LOAN_CLASSIFICATION, and ACTUAL_ROI data in your DataFrame
# For example, if you have a DataFrame called df:
# df = pd.read_csv('your_data.csv')  # Make sure to load your actual data

# Calculate the correlation matrix
correlation_matrix = df[['TU_SCORE', 'LOAN_CLASSIFICATION', 'ACTUAL_ROI']].corr()

# Plot a heatmap using Seaborn
plt.figure(figsize=(6, 4))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=.5)
plt.title('Correlation Heatmap')
plt.show()
```
